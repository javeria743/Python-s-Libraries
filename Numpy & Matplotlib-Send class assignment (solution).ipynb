#### The data set has COVID-19 fictional data from 3 planets: Tatooine, Naboo and Alderaan
For each Planet there is information for: Daily Cases, Recoveries and Deaths in this order for 4 cities

The dataset is created as lists below since file handling is not yet covered. Normally this sort of data is read into python from one or multiple files

The objective of this excercise is to teach you slicing, aggregating, plotting data in a numpy array


```python
# Tatooine Data (Planet 1)
tatooine_data = [
    # City 1
    [100, 80, 5],
    # City 2
    [120, 90, 4],
    #City 3
    [110, 85, 6],
    # City 4
    [115, 88, 3],
]

# Naboo Data (Planet 2)
naboo_data = [
    [90, 70, 4], 
    [110, 85, 3], 
    [100, 80, 5], 
    [95, 75, 2]
]


# Alderaan Data (Planet 3)
alderaan_data = [
    [85, 65, 3], 
    [105, 80, 2], 
    [95, 75, 4], 
    [90, 70, 1]
]

```

Make a 3D numpy array from this data
What is the shape of this array
Explain what this means

```python
import numpy as np

all_data = np.array([tatooine_data, naboo_data, alderaan_data])
np_array_3d = all_data

# Display the shape of the 3D array to confirm
print("Shape of the 3D NumPy array:",np_array_3d.shape)
```

Can you find out planet with the highest number of deaths?
- Scroll up to the definition and see where exactly the number of deaths are stored
- it is column at index 2 for each country
- try to access and print this column and use the aggregation function
- np.sum np.max

```python
# Extracting death data for each city on each planet
death_data = np_array_3d[:, :, 2]
print(death_data)
print("\n")
print(death_data.shape)
print("\n")

# Summing the deaths for each planet
total_deaths_per_planet = np.sum(death_data, axis=1)
print(total_deaths_per_planet)

# Finding the planet with the highest number of deaths
planet_names = ["Tatooine", "Naboo", "Alderaan"]
most_deaths = np.max(total_deaths_per_planet)
print("\n")

print("Highest number of deaths:", most_deaths)
print("\n")

index_of_most_deaths = np.argmax(total_deaths_per_planet)
planet_with_most_deaths = planet_names[index_of_most_deaths]

print("Planet with the most deaths:", planet_with_most_deaths)
```

Now try calculating average deaths per city (12 cities in total)
recall array.flatten()

```python
# Extracting death data for each city across all planets
death_data = np_array_3d[:, :, 2]  # This will be a 2D array (Planets x Cities)

# Flattening the array to get a single array of all death data
all_deaths = death_data.flatten()
print(all_deaths)
print("\n")

# Calculating the average deaths per city
average_deaths_per_city = np.mean(all_deaths)
print("Average deaths per city:", average_deaths_per_city)
```

Now can you create a bar chart of cases for each city on Tatooine

```python
import matplotlib.pyplot as plt
import numpy as np


# Assuming np_array_3d is the 3D array you have created

# Extracting the cases data for each city on Tatooine
cases_tatooine = np_array_3d[0, :, 0]  # This gives the cases for each city on Tatooine

# Creating the bar chart
plt.figure()
plt.bar(range(len(cases_tatooine)), cases_tatooine, color='blue')
plt.title('COVID-19 Cases in Each City on Tatooine')
plt.xlabel('City')
plt.ylabel('Number of Cases')
plt.xticks(range(len(cases_tatooine)), ['City 1', 'City 2', 'City 3', 'City 4'])
plt.show()
```

Now try creating sub plots for each planet

```python
import matplotlib.pyplot as plt
import numpy as np

# Assuming np_array_3d is the 3D array you have created

# Creating a figure with 3 subplots
fig, axes = plt.subplots(1, 3, figsize=(15, 5))  # 1 row, 3 columns

# Planet names
planet_names = ["Tatooine", "Naboo", "Alderaan"]

# Plotting data for each planet
for i, planet_name in enumerate(planet_names):
    cases_data = np_array_3d[i, :, 0]
    axes[i].bar(range(len(cases_data)), cases_data, color='blue')
    axes[i].set_title(f'COVID-19 Cases - {planet_name}')
    axes[i].set_xlabel('City')
    axes[i].set_ylabel('Number of Cases')
    axes[i].set_xticks(range(len(cases_data)))
    axes[i].set_xticklabels(['City 1', 'City 2', 'City 3', 'City 4'])

# Adjusting layout for better appearance
plt.tight_layout()

# Show the plot
plt.show()


```

Lets make the y axis consistent for comparison purpose

```python
import matplotlib.pyplot as plt
import numpy as np

# Assuming np_array_3d is the 3D array you have created

# Find the maximum number of cases across all cities and planets
max_cases = np.max(np_array_3d[:, :, 0])

# Creating a figure with 3 subplots
fig, axes = plt.subplots(1, 3, figsize=(18, 5))  # 1 row, 3 columns

# Planet names
planet_names = ["Tatooine", "Naboo", "Alderaan"]

# Plotting data for each planet
for i, planet_name in enumerate(planet_names):
    cases_data = np_array_3d[i, :, 0]
    axes[i].bar(range(len(cases_data)), cases_data, color='blue')
    axes[i].set_title(f'COVID-19 Cases - {planet_name}')
    axes[i].set_xlabel('City')
    axes[i].set_ylabel('Number of Cases')
    axes[i].set_xticks(range(len(cases_data)))
    axes[i].set_xticklabels(['City 1', 'City 2', 'City 3', 'City 4'])
    axes[i].set_ylim(0, max_cases)  # Set y-axis limit for consistency

# Adjusting layout for better appearance
plt.tight_layout()

# Show the plot
plt.show()
```
